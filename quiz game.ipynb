```python
from tkinter import *
from tkinter.ttk import Radiobutton,Combobox

quiz=Tk()
quiz.geometry('500x500')
quiz.configure(bg='#F7DED0')
quiz.resizable(height=False,width=False)
quiz.title("QUIZ GAME.py")
title=Label(quiz,text="QUIZZ GAME",font=('Verdana',14,'bold','italic'),pady=15,bg='#F7DED0')
title.pack()
resultt=0
var=IntVar()
def first():
    global frm1,resultt
    frm1=Frame(quiz,bg='#FEECE2')
    frm1.place(relx=.2,rely=.3,relheight=.6,relwidth=.66)
    ques=Label(frm1,text="Q1) In a website browser address bar,\n what does “www” stand for?",bg='#FEECE2',font=('arial',11,''))
    ques.place(relx=.01,rely=.01)
    opt1=Radiobutton(frm1,text='World web wide',value=1,variable=var)
    opt1.place(relx=.02,rely=.15)
    opt2=Radiobutton(frm1,text='Web Wide World',value=2,variable=var)
    opt2.place(relx=.02,rely=.25)
    opt3=Radiobutton(frm1,text='World Wide World',value=3,variable=var)
    opt3.place(relx=.02,rely=.35)
    opt4=Radiobutton(frm1,text='World Wide Web',value=4,variable=var)
    opt4.place(relx=.02,rely=.45)
    nextt=Button(frm1,text="Next Question",bd=2,font=('arial',10,''),command=second)
    nextt.place(relx=.75,rely=.9)
def second():
    global frm1,resultt
    if var.get() == 4:
        resultt+=1
    frm1.destroy()
    frm1=Frame(quiz,bg='#FEECE2')
    frm1.place(relx=.2,rely=.3,relheight=.6,relwidth=.66)
    ques=Label(frm1,text="Q2) In 2009, what became the first Morse\n code character to be added since WWII?",bg='#FEECE2',font=('arial',11,''))
    ques.place(relx=.01,rely=.01)
    opt1=Radiobutton(frm1,text='# symbol',value=1,variable=var)
    opt1.place(relx=.02,rely=.15)
    opt2=Radiobutton(frm1,text='@ symbol',value=2,variable=var)
    opt2.place(relx=.02,rely=.25)
    opt3=Radiobutton(frm1,text='$ symbol',value=3,variable=var)
    opt3.place(relx=.02,rely=.35)
    opt4=Radiobutton(frm1,text='? symbol',value=4,variable=var)
    opt4.place(relx=.02,rely=.45)
    nextt=Button(frm1,text="Next Question",bd=2,font=('arial',10,''),command=third)
    nextt.place(relx=.75,rely=.9)
def third():
    global frm1,resultt
    if var.get() == 2:
        resultt+=1
    frm1.destroy()
    frm1=Frame(quiz,bg='#FEECE2')
    frm1.place(relx=.2,rely=.3,relheight=.6,relwidth=.66)
    ques=Label(frm1,text="Q3) Which of one is not follows \nOOP's Concepts",bg='#FEECE2',font=('arial',11,''))
    ques.place(relx=.01,rely=.01)
    opt1=Radiobutton(frm1,text='C language',value=1)
    opt1.place(relx=.02,rely=.15)
    opt2=Radiobutton(frm1,text='Python',value=2)
    opt2.place(relx=.02,rely=.25)
    opt3=Radiobutton(frm1,text='Django',value=3)
    opt3.place(relx=.02,rely=.35)
    opt4=Radiobutton(frm1,text='Javascript',value=4)
    opt4.place(relx=.02,rely=.45)
    nextt=Button(frm1,text="Next Question",bd=2,font=('arial',10,''),command=fourth)
    nextt.place(relx=.75,rely=.9)
def fourth():
    global frm1,resultt
    if var.get() == 3:
        resultt+=1
    frm1.destroy()
    frm1=Frame(quiz,bg='#FEECE2')
    frm1.place(relx=.2,rely=.3,relheight=.6,relwidth=.66)
    ques=Label(frm1,text="Q4) What Does SQL stands for ??",bg='#FEECE2',font=('arial',11,''))
    ques.place(relx=.01,rely=.01)
    opt1=Radiobutton(frm1,text='Sturctured Data Locally',value=1)
    opt1.place(relx=.02,rely=.15)
    opt2=Radiobutton(frm1,text='Store Query Load',value=2)
    opt2.place(relx=.02,rely=.25)
    opt3=Radiobutton(frm1,text='Sturtured Query Language',value=3)
    opt3.place(relx=.02,rely=.35)
    opt4=Radiobutton(frm1,text='Simple Query Lnguage',value=4)
    opt4.place(relx=.02,rely=.45)
    nextt=Button(frm1,text="Next Question",bd=2,font=('arial',10,''),command=result)
    nextt.place(relx=.75,rely=.9)
def result():
    global frm1,resultt
    if var.get() == 3:
        resultt+=1
    frm1.destroy()
    frm1=Frame(quiz,bg='#FEECE2')
    frm1.place(relx=.2,rely=.3,relheight=.6,relwidth=.66)
    label=Label(frm1,text=f'your correct answers  {resultt} &\n& incorrect answers  {4-resultt}\n Thank You',font=('Arial',18,'italic'))
    label.config(bg='#FEECE2')
    label.place(relx=.1,rely=.35)
first()
quiz.mainloop()
```
