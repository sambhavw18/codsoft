```python
# calculator

from tkinter import *
def _0():
    global i
    i=0
    e.insert(i,'0')
    i+=1
def _add():
    e.insert(i+1,'+')
def mul():
    e.insert(i+1,'*')
def sub():
    e.insert(i+1,'-')
def div():
    e.insert(i+1,'/')
def _1():
    e.insert(i+1,'1')
def _2():
    e.insert(i+1,'2')
def _3():
    e.insert(i+1,'3')
def _4():
    e.insert(i+1,'4')
def _5():
    e.insert(i+1,'5')
def _6():
    e.insert(i+1,'6')
def _7():
    e.insert(i+1,'7')
def _8():
    e.insert(i+1,'8')
def _9():
    e.insert(i+1,'9')
def clear():
    e.delete(0,'end')
def result():
    a=e.get()
    b=eval(a)
    e.delete(0,'end')
    e.insert(0,b)
cal=Tk()
cal.geometry("250x350")


frm=Frame(cal)
frm.pack(side='bottom',anchor='c')

e=Entry(cal,text='')
e.pack(side='bottom',anchor='c')
btn=Button(frm,text='0',bg='grey',width=3,height=2,command=_0)
btn1=Button(frm,text='1',bg='grey',width=3,height=2,command=_1)
btn2=Button(frm,text='2',bg='grey',width=3,height=2,command=_2)
btn3=Button(frm,text='3',bg='grey',width=3,height=2,command=_3)
btn4=Button(frm,text='4',bg='grey',width=3,height=2,command=_4)
btn5=Button(frm,text='5',bg='grey',width=3,height=2,command=_5)
btn6=Button(frm,text='6',bg='grey',width=3,height=2,command=_6)
btn7=Button(frm,text='7',bg='grey',width=3,height=2,command=_7)
btn8=Button(frm,text='8',bg='grey',width=3,height=2,command=_8)
btn9=Button(frm,text='9',bg='grey',width=3,height=2,command=_9)
btnx=Button(frm,text='x',bg='grey',width=3,height=2,command=clear)
btne=Button(frm,text='=',bg='grey',width=3,height=2,command=result)
btnp=Button(frm,text='+',bg='grey',width=3,height=2,command=_add)
btns=Button(frm,text='-',bg='grey',width=3,height=2,command=sub)
btnm=Button(frm,text='*',bg='grey',width=3,height=2,command=mul)
btnd=Button(frm,text='/',bg='grey',width=3,height=2,command=div)



btn.grid(row=0,column=0)
btn1.grid(row=0,column=1)
btn2.grid(row=0,column=2)
btnp.grid(row=0,column=3)
btn3.grid(row=1,column=0)
btn4.grid(row=1,column=1)
btn5.grid(row=1,column=2)
btns.grid(row=1,column=3)
btn6.grid(row=2,column=0)
btn7.grid(row=2,column=1)
btn8.grid(row=2,column=2)
btnm.grid(row=2,column=3)
btn9.grid(row=3,column=0)
btnx.grid(row=3,column=1)
btne.grid(row=3,column=2)
btnd.grid(row=3,column=3)

cal.mainloop()
```
