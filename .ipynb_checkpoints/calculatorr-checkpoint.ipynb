```python
from tkinter import *
i=0
calc=Tk()
calc.geometry('400x400')
calc.config(bg='#E9E3E9')
calc.title("Calculator")
lb=Label(calc,text='Calculator',pady=30,bg="#E9E3E9",font=('arial',16,'bold'))
lb.pack()
main_frame = Frame(calc, bg='yellow')
main_frame.pack(side='bottom')
def num1():
    global i
    ent.insert(i,1)
    i+=1
def num2():
    global i
    ent.insert(i,2)
    i+=1
def num3():
    global i
    ent.insert(i,3)
    i+=1
def num4():
    global i
    ent.insert(i,4)
    i+=1
def num5():
    global i
    ent.insert(i,5)
    i+=1
def num6():
    global i
    ent.insert(i,6)
    i+=1
def num7():
    global i
    ent.delete(0,'end')
    ent.insert(i,7)
    i+=1
def num8():
    global i
    ent.insert(i,8)
    i+=1
def num9():
    global i
    ent.insert(i,9)
def plus():
    global i
    ent.insert(i,'+')
    i+=1
def sub():
    global i
    ent.insert(i,'-')
    i+=1
def mul():
    global i
    ent.insert(i,'*')
    i+=1
def div():
    global i
    ent.insert(i,'//')
    i+=1
def num0():
    global i
    ent.insert(i,0)
    i+=1
def equal():
    ques=ent.get()
    ans=eval(ques)
    ent.delete(0,'end')
    ent.insert(0,ans)
def clearr():
    ent.delete(0,'end')
ent=Entry(calc,font=('arial',14,''),justify="right",width=12)
ent.pack(side='bottom')
btn1=Button(main_frame,text="1",font=('devangiri',10,''),width=3,height=2,command=num1,bg="#d5bdaf")
btn1.grid(row=0,column=0)
btn2=Button(main_frame,text='2',font=('devangiri',10,''),width=3,height=2,command=num2,bg="#d5bdaf")
btn2.grid(row=0,column=1)
btn3=Button(main_frame,text='3',font=('devangiri',10,''),width=3,command=num3,height=2,bg="#d5bdaf")
btn3.grid(row=0,column=2)
btnp=Button(main_frame,text='+',font=('devangiri',10,''),width=3,command=plus,height=2,bg="#d5bdaf")
btnp.grid(row=0,column=3)
btn4=Button(main_frame,text='4',font=('devangiri',10,''),width=3,command=num4,height=2,bg="#d5bdaf")
btn4.grid(row=1,column=0)
btn5=Button(main_frame,text='5',font=('devangiri',10,''),width=3,command=num5,height=2,bg="#d5bdaf")
btn5.grid(row=1,column=1)
btn6=Button(main_frame,text='6',font=('devangiri',10,''),width=3,command=num6,height=2,bg="#d5bdaf")
btn6.grid(row=1,column=2)
btns=Button(main_frame,text='-',font=('devangiri',10,''),width=3,command=sub,height=2,bg="#d5bdaf")
btns.grid(row=1,column=3)
btn7=Button(main_frame,text='7',font=('devangiri',10,''),width=3,command=num7,height=2,bg="#d5bdaf")
btn7.grid(row=2,column=0)
btn8=Button(main_frame,text='8',font=('devangiri',10,''),width=3,command=num8,height=2,bg="#d5bdaf")
btn8.grid(row=2,column=1)
btn9=Button(main_frame,text='9',font=('devangiri',10,''),width=3,command=num9,height=2,bg="#d5bdaf")
btn9.grid(row=2,column=2)
btnm=Button(main_frame,text='x',font=('devangiri',10,''),width=3,command=mul,height=2,bg="#d5bdaf")
btnm.grid(row=2,column=3)
clear=Button(main_frame,text='clear',font=('devangiri',10,''),bg="#EF8275",width=3,command=clearr,height=2)
clear.grid(row=3,column=0)
cros=Button(main_frame,text='0',font=('devangiri',10,''),width=3,command=num0,height=2,bg="#d5bdaf")
cros.grid(row=3,column=1)
btneq=Button(main_frame,text='=',font=('devangiri',10,''),width=3,bg="#d5bdaf",command=equal,height=2)
btneq.grid(row=3,column=2)
btndi=Button(main_frame,text='/',font=('devangiri',10,''),width=3,bg="#d5bdaf",command=div,height=2)
btndi.grid(row=3,column=3)
calc.mainloop()
```
