```python
from tkinter import *
from tkinter import messagebox

def addonclick():
    a=en.get()
    b=en1.get()
    c=int(a)+int(b)
    result.delete(0,'end')
    result.insert(0,c)

def mulonclick():
    a=en.get()
    b=en1.get()
    c=int(a)*int(b)
    result.delete(0,'end')
    result.insert(0,c)    
    
def subonclick():
    a=en.get()
    b=en1.get()
    c=int(a)-int(b)
    result.delete(0,'end')
    result.insert(0,c)
    
def divonclick():
    a=en.get()
    b=en1.get()
    c=int(a)//int(b)
    result.delete(0,'end')
    result.insert(0,c)

def squareonclick():
    a=enter.get()
    c=(int(a))**2
    res.delete(0,'end')
    res.insert(0,c)
    
def cubeonclick():
    a=enter.get()
    c=(int(a))**3
    res.delete(0,'end')
    res.insert(0,c)
    
disp=Tk()
disp.geometry('600x500')
disp.configure(bg="white")

num=Label(disp,text='enter two numbers')
num.place(x=320,y=220)

en=Entry(disp)
en.place(x=320,y=250)
en1=Entry(disp)
en1.place(x=320,y=270)

add=Button(disp,text='+',width=2,height=2,bg='orange',bd=4,font=('arial',10,'bold'),command=addonclick)
sub=Button(disp,text='-',width=2,height=2,bg='orange',bd=4,font=('arial',10,'bold'),command=subonclick)
mul=Button(disp,text='x',width=2,height=2,bg='orange',bd=4,font=('arial',10,'bold'),command=mulonclick)
div=Button(disp,text='/',width=2,height=2,bg='orange',bd=4,font=('arial',10,'bold'),command=divonclick)

la=Label(disp,text='Result')
la.place(x=320,y=350)
result=Entry(disp,bd=2)
result.place(x=320,y=380)

add.place(x=550,y=300)
sub.place(x=550,y=350)
mul.place(x=550,y=400)
div.place(x=550,y=450)

sq=Button(disp,text='a²',font=('arial',9,'bold'),command=squareonclick)
sq.place(x=50,y=360)
cube=Button(disp,text='a³',font=('arial',9,'bold'),command=cubeonclick)
cube.place(x=80,y=360)
enter=Entry(disp)
enter.place(x=50,y=330)

output=Label(disp,text='Result')
output.place(x=10,y=390)
res=Entry(disp,bd=2)
res.place(x=50,y=390)
disp.mainloop()
```
