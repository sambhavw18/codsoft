# Python Quiz

```python
from tkinter import *
from tkinter import messagebox

def result():
    Result.insert(0,points)
    messagebox.showinfo('result',f"Correct Answers :{points} Incorrect Answers :{4-points}")
def answer():
    global points
    points=0
    a=var.get()
    if a == 2:
        points+=1
    a1=var1.get()
    if a1==3:
        points+=1
    a2=var2.get()
    if a2 == 4:
        points+=1
    a3=var3.get()
    if a3 == 1:
        points+=1
    result()
scr=Tk()

croll=Scrollbar(scr)
croll.pack(side='right')
#scr(yscroolcommand=croll.set)

scr.title('Code With Sambhav')
scr.geometry('500x600')
Label(scr,text="Python Quiz",font=('new times roman',10,'bold'),pady=10).pack()

ques1=Label(scr,text='Q1 : Who created python ?')
ques1.pack(anchor='w')
var=IntVar()
opt1=Radiobutton(scr,text='Chris Jordan',variable=var,value=1)
opt1.pack(anchor='w')
opt1=Radiobutton(scr,text='Guido Van Rossum',variable=var,value=2)
opt1.pack(anchor='w')
opt1=Radiobutton(scr,text='Dennis Ritchie',variable=var,value=3)
opt1.pack(anchor='w')
opt1=Radiobutton(scr,text='James Gosling',variable=var,value=4)
opt1.pack(anchor='w')

ques2=Label(scr,text='Q2 : What is a correct syntax to output "Hello World" in Python?')
ques2.pack(anchor='w')
var1=IntVar()
opt2=Radiobutton(scr,text='printf(Hello World)',variable=var1,value=1)
opt2.pack(anchor='w')
opt2=Radiobutton(scr,text='echo("Hello World")',variable=var1,value=2)
opt2.pack(anchor='w')
opt2=Radiobutton(scr,text='print("Hello World")',variable=var1,value=3)
opt2.pack(anchor='w')
opt2=Radiobutton(scr,text="p('Hello World')",variable=var1,value=4)
opt2.pack(anchor='w')

ques3=Label(scr,text='Q3 : What is the correct way to create a function in Python?')
ques3.pack(anchor='w')
var2=IntVar()
opt3=Radiobutton(scr,text='create myfunc()',variable=var2,value=1)
opt3.pack(anchor='w')
opt3=Radiobutton(scr,text='Def myfunc()',variable=var2,value=2)
opt3.pack(anchor='w')
opt3=Radiobutton(scr,text='function myfunc()',variable=var2,value=3)
opt3.pack(anchor='w')
opt3=Radiobutton(scr,text="def myfunc():",variable=var2,value=4)
opt3.pack(anchor='w')

ques4=Label(scr,text='Q4 : Which of these defines a Empty LIST?')
ques4.pack(anchor='w')
var3=IntVar()
opt4=Radiobutton(scr,text="[]",variable=var3,value=1)
opt4.pack(anchor='w')
opt4=Radiobutton(scr,text='{}',variable=var3,value=2)
opt4.pack(anchor='w')
opt4=Radiobutton(scr,text='()',variable=var3,value=3)
opt4.pack(anchor='w')
opt4=Radiobutton(scr,text="['','','']",variable=var3,value=4)
opt4.pack(anchor='w')
Button(scr,text='submit',command=answer).pack()



Result=Entry(scr)
Result.pack(side='bottom')
scr.mainloop()
```
